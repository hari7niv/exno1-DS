```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
age=[1,3,28,27,25,92,30,39,40,50,26,24,29,94]
dt = pd.DataFrame(age)
dt
```

```python
sns.boxplot(data = dt)
```

```python
sns.scatterplot(data = dt)
```

```python
q1= dt.quantile(0.25)
q3= dt.quantile(0.75)
iqr = q3-q1
iqr
```

```python
Q1=np.percentile(dt,25)
Q3=np.percentile(dt,75)
IQR=Q3-Q1
IQR
```

```python

lb = Q1-1.5*IQR
lb
```

```python
ub = Q3 +1.5 *IQR
ub
```

```python
outliers=[x for x in age if x<lb or x>ub]
print("Q1:",Q1)
print("Q3:",Q3)
print("IQR:",IQR)
print("Lower Bound:",lb)
print("Upper Bound:",ub)
print("Outliers:",outliers)
```

```python
dt = dt[((dt>=lb)&(dt<=ub))]
dt
```

```python
dt.dropna()
```

```python
sns.boxplot(data=dt)
```

```python
data=[1,2,2,2,3,1,1,15,2,2,2,3,1,1,2]
mean=np.mean(data)
std=np.std(data)
print('mean of the dataset is',mean)
print('std.deviation is',std)
```

```python
threshold=3
outlier=[]
for i in data:
  z=(i-mean)/std
  if z>threshold:
    outlier.append(i)
print('outlier in dataset is',outlier)
```
